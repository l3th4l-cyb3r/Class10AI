#### Python program to merge two images

```python
from PIL import Image
image = Image.open("ocean.jpg")
r,g,b = image.split()
image.show()
image = Image.merge("RGB", (b,g,r))
image.show()
```

#### Python program to convert image from RGB to Grayscale

```python
from PIL import Image
img = Image.open("ocean.jpg")
img.convert('1')
```

# Python program to display recently updated data

```python
import pandas as pd
data_sorted = pd.read_csv('data-file.csv')
data_sorted.sort_values([data_sorted.columns[1]],
axis=0,
ascending=[False],
inplace=True)
display(data_sorted)
```

# Program to plot a bar chart to display 

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

data = pd.read_csv('data-file.csv')
data['Power'].value_counts().sort_index().plot(kind='bar')
plt.show()
```
